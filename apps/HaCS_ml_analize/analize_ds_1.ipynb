```python
import pandas as pd
from openpyxl import load_workbook
from typing import Iterable
from pydantic import BaseModel, ConfigDict
pd.set_option('display.max_columns', None)
```

# **Classes**

```python
class SecondaryTable(BaseModel):
    model_config=ConfigDict(arbitrary_types_allowed=True)
    sheet_name: str
    describe_title: str
    table: pd.DataFrame
```

# **Functions**

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

```python
def load_secondary_tables(path: str, secondary_tables: Iterable):
    for st in secondary_tables:
        df = pd.read_excel(path, engine='openpyxl', sheet_name=st)
        df_p = df.iloc[1:]
        df_p.columns = df.iloc[0].list()
        describe_title = df.columns[0]
        SecondaryTable(
            sheet_name=st,
            describe_title=describe_title,
            
        )
```
