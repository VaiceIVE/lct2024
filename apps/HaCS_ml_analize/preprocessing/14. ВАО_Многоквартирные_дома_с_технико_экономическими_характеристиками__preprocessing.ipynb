# **Imports**

```python
import pandas as pd
from openpyxl import load_workbook
from dataclasses import dataclass
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
pd.set_option('display.max_columns', None)
```

# **Functions**

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

```python
def get_description_table(col_name):
    table = pd.read_excel(Pathes.path, engine='openpyxl', sheet_name=col_name)
    table.columns = table.iloc[0]
    table = table.iloc[1:]
    idx = table.columns[0]
    table.index = table[idx]
    return table[table.columns[1]].to_dict()
```

# **Constants**

```python
@dataclass(frozen=True)
class Pathes:
    path: str = '../main_datasets/10. ДЖКХ + ДИТ/14. ВАО_Многоквартирные_дома_с_технико_экономическими_характеристиками.xlsx'
    path_processed: str = '../main_datasets/processed_datasets/'
```

# **Load dataset**

```python
sheets_list = get_sheetnames_xlsx(Pathes.path)
sheets_list
```

## ВАО

```python
table = pd.read_excel(Pathes.path, engine='openpyxl', sheet_name=sheets_list[0])
```

```python
table = table.drop('col_1945_del', axis=1)
```

```python
table
```

```python
true_column = table.iloc[0]
table = table.iloc[1:]
table.columns = true_column
```

```python
for i in tqdm(sheets_list[1:]):
    d = get_description_table(i)
    if i.lower() in table.columns:
        table[i.lower()] = table[i.lower()].apply(lambda a: a if np.isnan(a) else d[a])
```

```python
table.columns = true_column
```

```python
table.to_csv(Pathes.path_processed + '14. ВАО_Многоквартирные_дома_с_технико_экономическими_характеристиками__ВАО.csv', index=False)
```
