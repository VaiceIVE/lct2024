# **Imports**

```python
import pandas as pd
from openpyxl import load_workbook
from dataclasses import dataclass
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

# **Functions**

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

# **Constants**

```python
@dataclass(frozen=True)
class Pathes:
    path: str = '../main_datasets/10. ДЖКХ + ДИТ/11.Выгрузка_ОДПУ_отопление_ВАО_20240522.xlsx'
    path_processed: str = '../main_datasets/processed_datasets/'
```

# **Load dataset**

```python
sheets_list = get_sheetnames_xlsx(Pathes.path)
sheets_list
```

## Sheet 1

```python
table = pd.read_excel(Pathes.path, engine='openpyxl', sheet_name=sheets_list[0])
```

```python
table.dropna(axis = 0, how = 'all')
```

```python
table.to_csv(Pathes.path_processed + '11.Выгрузка_ОДПУ_отопление_ВАО_20240522__sheet1.csv', index=False)
```

## Sheet 2

```python
table2 = pd.read_excel(Pathes.path, engine='openpyxl', sheet_name=sheets_list[1])
```
