# **Imports**

```python
import pandas as pd
from openpyxl import load_workbook
from dataclasses import dataclass
import matplotlib.pyplot as plt
```

# **Functions**

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

# **Constants**

```python
@dataclass(frozen=True)
class Pathes:
    path: str = '../main_datasets/10. ДЖКХ + ДИТ/13. Адресный реестр объектов недвижимости города Москвы.xlsx'
    path_processed: str = '../main_datasets/processed_datasets/'
```

# **Load dataset**

```python
sheets_list = get_sheetnames_xlsx(Pathes.path)
sheets_list
```

```python
%time
table = pd.read_excel(Pathes.path, engine='openpyxl', sheet_name=sheets_list[0])
```

```python
c = table.iloc[0]
table_ = table.iloc[1:]
table_.columns = c
```

```python
table_
```

```python
table_.to_csv(Pathes.path_processed + '13. Адресный реестр объектов недвижимости города Москвы____Processed.csv', index=False)
```
