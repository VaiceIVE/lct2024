# **Imports**

```python
import pandas as pd
from dataclasses import dataclass
from openpyxl import load_workbook
pd.set_option('display.max_columns', None)
```

# **Constants**

```python
@dataclass(frozen=True)
class Pathes:
    odpu_heating_uploading_sheet1: str = '../main_datasets/processed_datasets/11.Выгрузка_ОДПУ_отопление_ВАО_20240522__sheet1.csv'
    odpu_heating_uploading_sheet2: str = '../main_datasets/processed_datasets/11.Выгрузка_ОДПУ_отопление_ВАО_20240522__sheet2.csv'

```

# **Functions**

```python
def replace_bad_spaces(df: pd.DataFrame) -> pd.DataFrame:
    c = df.columns.map(lambda a: str(a).rstrip().lstrip())
    df.columns = c
    return df
```

# **Concat**

```python
dfs = [replace_bad_spaces(pd.read_csv(Pathes.__getattribute__(Pathes, i))) for i in Pathes.__annotations__.keys()]
```

```python
table = pd.concat(dfs, axis=0)
```
