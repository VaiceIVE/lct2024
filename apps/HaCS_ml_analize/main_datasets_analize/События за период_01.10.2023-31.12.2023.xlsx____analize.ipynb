# **Imports**

```python
import pandas as pd
from openpyxl import load_workbook
from dataclasses import dataclass
import matplotlib.pyplot as plt
```

# **Functions**

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames
```

# **Constants**

```python
@dataclass(frozen=True)
class Pathes:
    path: str = '../main_datasets/processed_datasets/События за период_01.10.2023-31.12.2023.xlsx____uploading_table.csv'
    
```

# **Analize**

```python
data = pd.read_csv(Pathes.path)
print(data.columns)
names = data['Наименование'].value_counts().to_dict()
sources = data['Источник'].value_counts().to_dict()
```

```python
plt.figure(figsize=(30, 5))
plt.xticks(rotation=90)
plt.bar(names.keys(), names.values(), width=0.8)
```

```python
plt.xticks(rotation=90)
plt.bar(sources.keys(), sources.values(), width=0.8)
```
